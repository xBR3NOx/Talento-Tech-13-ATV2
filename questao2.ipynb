```python
import matplotlib.pyplot as plt
import numpy as np

# Criando os horários de 0h a 24h (eixo X)
horarios = np.arange(0, 25, 1)  # De 0 a 24

# Criando uma curva fictícia para a temperatura ao longo do dia (eixo Y)
temperaturas = [15 + (i / 2) if i <= 12 else 30 - ((i - 12) / 2) for i in horarios]

# Criando o gráfico
plt.figure(figsize=(10, 5))
plt.plot(horarios, temperaturas, marker='o', linestyle='-', color='b', label="Temperatura")

# Personalizando o gráfico
plt.title("Evolução da Temperatura ao Longo do Dia")
plt.xlabel("Horário (h)")
plt.ylabel("Temperatura (°C)")
plt.xticks(range(0, 25, 2))  # Exibir os horários de 2 em 2 horas
plt.grid(True)
plt.legend()

# Exibir o gráfico
plt.show()
```
